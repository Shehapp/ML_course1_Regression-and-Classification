```python
import numpy as np
import seaborn as sns
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,3])
print(f"this is e^x {np.exp(x1)}") # exp(x) e^x
```

```python
#sigmoid function
def sigmoid(y_hat):
    return 1/(1+np.exp(-y_hat))
```

```python
z = np.arange(-10,11)
z_sigm = sigmoid(z)
print(np.c_[z,z_sigm])
```

```python
plt.scatter(x=z,y=z_sigm,c='g')
plt.title("Z vs Sigmoid(Z)")
plt.xlabel("z")
plt.ylabel("Sigmoid(Z)")
plt.show()
```
