```python
#import numpy and Matplotlib(plot)
import numpy as np
import matplotlib.pyplot as plt
```

```python
#training sets
x_train=np.array([1,2])
y_train=np.array([300,500])

#print it
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
#number of training examples (m)
m = x_train.shape[0]
print(f"m = {m}")
#shape of data (n,m) change it => x_train.reshape(n,m)
print(f"shape of data: {x_train.shape}")
```

```python
#ploting the data
#          x-axis   y-axis   shape     color
plt.scatter(x_train,y_train,marker='*',c='r')
#plt.plot(x_train,y_train,marker='*',c='r') # fit straight line
#titel
plt.title("housing price")
#set x-axis label
plt.xlabel("feet")
#set y-axis label
plt.ylabel("price")
plt.show()

```

```python
#fun take b,w and feature and return prediction
def compute_prediction(x,w,b):
    # f(x)=x*w+b
    f_wb=[]
    for i in x:
        f_wb.append(i*w+b)
    return f_wb
```

```python
#let's start with w=100 and b=100
w=100
b=100
print(f"w = {w}")
print(f"b = {b}")
```

```python
# compute perdict
y_hat=compute_prediction(x_train,w,b)

#Plot our model prediction
plt.plot(x_train,y_hat,c='g',label="Our prediction")
#plot training examples
plt.scatter(x_train,y_train,c='r',label="Actual values")
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()

#output is very bad w,b is not good try values 
# find if w=200 and x=100 is the best fit line


```

```python
w=200
b=100
# compute perdict
y_hat=compute_prediction(x_train,w,b)

#Plot our model prediction
plt.plot(x_train,y_hat,c='g',label="Our prediction")
#plot training examples
plt.scatter(x_train,y_train,c='r',label="Actual values")
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# this is the model ouput the best line
def result(feature):
    b=100
    w=200
    return feature*w+b
```
